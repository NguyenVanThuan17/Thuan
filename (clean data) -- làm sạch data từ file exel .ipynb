```python
import numpy as np 
import pandas as pd
```

```python
# đọc data từ file exel 
df1=pd.read_excel('E:\dataset\Sample_Superstore.xls' , sheet_name='Missing') 
df1
```

```python
df1.head()
```

```python
#check is there any  missing data --- check cái NaN trong bảng data
for i in df1.columns:
        miss= df1[i].isnull().sum()
        if miss > 0 : 
            print(f'{i}  column has {miss} missing data')
        else : 
            print(f'{i} column has no missing data')
```

```python
df1.isnull().sum()
```

```python
#fill mising data -- thay những giá trị NaN bằng bfill or ffill
df1['Product'].fillna(method='bfill' , inplace = True)
df1['Sales'].fillna(method='bfill' , inplace = True)
df1['Quantity'].fillna(method='bfill' , inplace = True)
df1['Profit'].fillna(method='bfill' , inplace = True)
df1
```

```python
# drop missing data colum , row -- xóa những data NaN
df1.dropna()
```
